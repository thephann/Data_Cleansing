# Data cleansing

```python
fifa21_data.isnull().sum()
```

```python
fifa21_data['Volleys'].fillna(fifa21_data['Volleys'].mean(), inplace = True) 
fifa21_data['Curve'].fillna(fifa21_data['Curve'].mean(), inplace = True)
fifa21_data['Agility'].fillna(fifa21_data['Agility'].mean(), inplace = True)
fifa21_data['Balance'].fillna(fifa21_data['Balance'].mean(), inplace = True)
fifa21_data['Jumping'].fillna(fifa21_data['Jumping'].mean(), inplace = True)
fifa21_data['Interceptions'].fillna(fifa21_data['Interceptions'].mean(), inplace = True)
fifa21_data['Positioning'].fillna(fifa21_data['Positioning'].mean(), inplace = True)
fifa21_data['Composure'].fillna(fifa21_data['Composure'].mean(), inplace = True)
fifa21_data['Sliding Tackle'].fillna(fifa21_data['Sliding Tackle'].mean(), inplace = True) 
fifa21_data['Vision'].fillna(fifa21_data['Vision'].mean(), inplace = True) 
```

```python
def value_handling(Value):
    if isinstance(Value,str):
        out = Value.replace('€', '')
        if 'M' in out:
            out = float(out.replace('M', ''))*1000000
        elif 'K' in Value:
            out = float(out.replace('K', ''))*1000
        return float(out)
```

```python
fifa21_data['Value'] = fifa21_data['Value'].apply(lambda x: value_handling(x))
fifa21_data['Release Clause'] = fifa21_data['Release Clause'].apply(lambda x: value_handling(x))
fifa21_data['Release Clause'].fillna(fifa21_data['Release Clause'].mean(), inplace = True)
```

```python
fifa21_data.head(5)
```

```python
def wage_handling(value):
    out = value.replace('€', '')
    if 'K' in out:
        out = float(out.replace('K', ''))*1000
    else:
        out = out
    return float(out)

fifa21_data['Wage'] = fifa21_data['Wage'].apply(lambda x: wage_handling(x))
fifa21_data.Wage.head(3)
```

```python
# Height columns
fifa21_data.Height.head(3)
```

```python
def feet_to_cm(value):
    h_inch = int(value.split("'")[1][:-1])
    h_inch += int(value.split("'")[0]) * 12
    h_cm = round(h_inch * 2.54, 1)
    return h_cm

fifa21_data['Height'] = fifa21_data['Height'].apply(lambda x: feet_to_cm(x))
fifa21_data['Height'].fillna((fifa21_data['Height'].mean()), inplace = True)
fifa21_data.Height.head(5)
```

```python
# Weight columns
def lbs_to_kg(x):
    out = float(x.replace('lbs', ''))
    kgs = out/2.2046
    out = round(kgs , 2)
    return float(out)

fifa21_data['Weight'] = fifa21_data['Weight'].apply(lambda x: lbs_to_kg(x))
fifa21_data['Weight'].fillna((fifa21_data['Weight'].mean()), inplace = True)
fifa21_data.Weight.head(5)
```

```python
def skill_rating(value):
    if isinstance(value,str):
        s = int(value[0:2]) + int(value[-1])
        return s
    else:
        return value
```

```python
skill_columns = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM',
       'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM',
       'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']

for col in skill_columns:
    fifa21_data[col] = fifa21_data[col].apply(skill_rating)
```

```python
def skill_rating_1(value):
    val_split = value.split('+')
    out = int(val_split[0]) + int(val_split[1])
    return out
    
fifa21_data['GK'] = fifa21_data['GK'].apply(skill_rating_1)
```

```python
def foot_conversion(value):
    if (value == 'Right'):
        return 1
    else:
        return 0
    
fifa21_data['foot'] = fifa21_data['foot'].apply(foot_conversion)
```

```python
AW_DW = {
    'High': 2,
    'Medium': 1,
    'Low': 0
}
fifa21_data['A/W'] = fifa21_data['A/W'].map(AW_DW)
fifa21_data['D/W'] = fifa21_data['D/W'].map(AW_DW)
```

```python
fifa21_data.BP.unique()
```

```python
def BP_conversion(value):
    if (value == 'GK'):
        return 'GK'
    elif ((value == 'LWB') | (value == 'CB') | (value == 'RWB') | (value == 'LB') | (value == 'RB')):
        return 'DF'
    elif ((value == 'CDM')):
        return 'DM'
    elif ((value == 'LM') | (value == 'CM') | (value == 'RM')):
        return 'MF'
    elif ((value == 'CAM')):
        return 'AM'
    elif ((value == 'RW') | (value == 'ST') | (value == 'LW') | (value == 'CF')):
        return 'ST'
    else:
        return value

fifa21_data['BP'] = fifa21_data['BP'].apply(BP_conversion)
```

```python
BP_position = {
    'GK':0,
    'DF':1,
    'DM':2,
    'MF':3,
    'AM':4,
    'ST':5
}
fifa21_data['BP'] = fifa21_data['BP'].map(BP_position)
```
